```python
from fuzzywuzzy import process
```

```python
country_category=['USA','United States','Canada','America','UK','United Kingdom'] # the most common country
```

```python
from fuzzywuzzy import process
for state in country_category:
# Find potential matches in states with typoes
    matches = process.extract(state, candy_data['country'], limit = candy_data.shape[0])
# For each potential match match
    for potential_match in matches:
# If high similarity score
        if potential_match[1] >= 80:
# Replace typo with correct category
            candy_data.loc[candy_data['country'] == potential_match[0], 'country'] = state
```

```python
candy_data['country'].value_counts()
```

> we can see that the number of usa had increased since we replaced same of usa with different wrinting way to usa

```python
candy_data['country'].unique()
```

> i will try replace the rest of USA and UK that the match function fail to get

```python
for x in candy_data['country']:
    if ((('U' in str(x) and 'S' in str (x)))):
        candy_data['country']=candy_data['country'].replace(x,"USA")
```

```python
candy_data['country']=candy_data['country'].replace(['United States','America','New York','MEXICO','CALIFORNIA','NEW JERSEY','U.S.A.','US OF A','U S A'],'USA')
candy_data['country']=candy_data['country'].replace(['United Kingdom','ENGLAND','ENDLAND'],'UK')
candy_data['country'].value_counts()
```

```python
candy_data['country'].unique()
```

```python
# i have so many country that apperes only once so i should map them all to new category
#so i will remove all the unique values that are more that 5
other = [x for x in candy_data['country'].unique()]
other.remove('USA')
other.remove('Canada')
other.remove('UK')
other.remove('GERMANY')
other.remove('NETHERLANDS')
other.remove('JAPAN')
other.remove('SCOTLAND')
print(other)
```

```python
candy_data['country']=candy_data['country'].replace(other,'Other')
```
